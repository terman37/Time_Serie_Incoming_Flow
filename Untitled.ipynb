```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels as sm
```

```python
filename = './data/ML_IGE_ENTRANTS_MODELE2.csv'
```

```python
df = pd.read_csv(filename,sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df['SITE']=df['SITE'].astype('category')
df['SERVICE_ACTIVITE']=df['SERVICE_ACTIVITE'].astype('category')
df['DATEDATA']=pd.to_datetime(df['DATEDATA'])
```

```python
df.rename(columns={'SERVICE_ACTIVITE':'SERVICE','DATEDATA':'DATE'},inplace=True)
```

```python
df.head()
```

```python
df2 = df.groupby('DATE').sum()
```

```python
df2.plot()
```

```python
df2.info()
```

```python
df3 = df2[df2.index<'2019-01-01']
```

```python
df3.plot()
```

```python
data = df3.copy()

from statsmodels.tsa.tsatools import detrend

notrend = detrend(data.TOTAL)
data['notrend']=notrend
```

```python
data.plot()
```

```python
from statsmodels.tsa.stattools import acf
cor = acf(data.notrend,fft=False)
plt.plot(cor)
```

```python
from statsmodels.tsa.stattools import pacf
pcor = pacf(data.notrend)
plt.plot(pcor)
```

```python
plt.acorr(data.notrend)
```
